```python
X = [[1,2,3,4], [1,3,0,1], [1,2,1,1], [0,3,0,1]] # on cr√©e une matrice 4x4
```

```python
B  = [1,1,1,1] # on cr√©e un vecteur de taille 4
```

```python
np.dot(X,B) # on multiplie la matrice X par le vecteur B
```

```python
A = np.linalg.solve(X,B) # on r√©sout le syst√®me lin√©aire XA = B
```

```python
np.dot(X,A) # on v√©rifie que le produit de X par A est bien √©gal √† B
```
