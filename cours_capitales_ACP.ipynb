```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

# ACP (Analyse en Composantes Principales)

```python
capitales = pd.read_csv('./data/capitales.csv', sep=';', index_col='Villes')
capitales = capitales.iloc[:, range(0, 12)]
capitales
```

```python
# PCA : permet de savoir quelles colonnes sont les plus importantes / permet de lier les colonnes entre elles
from sklearn.decomposition import PCA
model = PCA(svd_solver='full')
capitales_pca = model.fit_transform(capitales)
capitales_pca = pd.DataFrame(capitales_pca)
capitales_pca
```

```python
# Affiche les colonnes les plus importantes
# attention : les colonnes 1 et 2 de PCA ne correspondent pas aux colonnes 1 et 2 de capitales. Ce sont des combinaisons linéaires des colonnes de capitales
model.explained_variance_ratio_
# model.explained_variance_
capitales_acp = pd.DataFrame(model.explained_variance_ratio_)

# On pourra supprimer les colonnes les moins importantes pour reduire la taille et supprimer le bruit

```

```python
# supprimer deux colonnes
capitales_pca = capitales_pca.iloc[:, :2]
capitales_pca
```

# CAH (Classification Ascendante Hiérarchique)

```python
from scipy.cluster.hierarchy import dendrogram, linkage
capitales_cah = linkage(capitales, method="ward", metric="euclidean")
```

```python
plt.title("CAH")
dendrogram(capitales_cah, labels=capitales.index, orientation="right", color_threshold=0)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(distance_threshold=None, n_clusters=5)
capitales_cah = model.fit(capitales)
```

```python
capitales_cah.labels_
```

```python
CAH = pd.DataFrame({"label": capitales_cah.labels_}, index=capitales.index)
CAH
```

# CAH with PCA (acp)

# K-means

```python
from sklearn.cluster import KMeans
inertie = []
for k in range(1,20) : 
  model_kmean = KMeans(n_clusters=k)
  model_kmean.fit(capitales)
  inertie.append(model_kmean.inertia_)
plt.plot(range(1,20), inertie)
```
