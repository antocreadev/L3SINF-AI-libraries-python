```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 3]

plt.plot(x, y)
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
categories = ['A', 'B', 'C']
values_1 = [10, 15, 12]
values_2 = [8, 17, 14]

plt.plot(categories, values_1, marker='o', label='2023')
plt.plot(categories, values_2, marker='o', label='2024')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Slopegraph')
plt.legend()
plt.show()
```

```python
G = nx.cycle_graph(10) 
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_size=10, node_color='skyblue')
plt.title('Network Graph')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 20]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel = 'Categories'
plt.ylabel = 'Values'

lchart = plt.twinx()
cumulative_values = [sum(values[:i+1]) for i in range(len(values))]
lchart.plot(categories, cumulative_values, color='red', marker='o')
lchart.set_ylabel('Cumulative values', color='red')
lchart.tick_params('y', colors='red')

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title = 'Histogram'
plt.xlabel = 'Values'
plt.ylabel = 'Frequency'
plt.show()
```

```python
x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)

plt.scatter(x, y)
plt.title = 'Scatter Plot'
plt.xlabel = 'X'
plt.ylabel = 'Y'
plt.show()
```
