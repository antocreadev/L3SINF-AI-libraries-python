```python
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
penguins = pd.read_csv("penguins.csv", sep=";")
penguins
```

```python
nan_individuals = penguins[penguins.isna().any(axis=1)]
nan_individuals
```

```python
indices_to_delete = [3, 271]
penguins = penguins.drop(indices_to_delete)
penguins
```

```python
data = penguins[["body_mass_g", "sex"]]
data
```

```python
moyennePoidsMale = data[data["sex"] == "male"]["body_mass_g"].mean()
moyennePoidsMale
```

```python
moyennePoidsFemelle = data[data["sex"] == "female"]["body_mass_g"].mean()
moyennePoidsFemelle
```

```python
sex_manquant = penguins[penguins["sex"].isna()]
sex_manquant
```

```python
penguins["sex"].replace({"female": 0, "male": 1}, inplace=True)

for index, row in sex_manquant.iterrows():
    poids = row["body_mass_g"]
    
    if abs(poids - moyennePoidsFemelle) < abs(poids - moyennePoidsMale):
        penguins.at[index, "sex"] = 0  # Female
    else:
        penguins.at[index, "sex"] = 1  # Male

penguins.loc[sex_manquant.index, "sex"] = penguins.loc[sex_manquant.index, "sex"].astype(int)
penguins
```

```python
iles = penguins["island"].unique()
iles
```

```python
especes = penguins["species"].unique()
especes
```

```python
island_codes = {'Torgersen': 0, 'Biscoe': 1, 'Dream': 2}
penguins['island'] = penguins['island'].replace(island_codes)
penguins
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(penguins, test_size=0.3, random_state=42)
print("Train set shape:", train_set.shape)
print("Test set shape:", test_set.shape)
```

```python
from sklearn.svm import SVC

x = penguins.drop("species", axis=1)
y = penguins["species"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = SVC(C=100)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

confusion_matrice = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_matrice, annot=True, xticklabels=model.classes_, yticklabels=model.classes_)
```

```python
model = SVC(kernel='linear', C=100)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
model.score(x_test, y_test)
```

```python
confusion_matrice2 = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_matrice2, annot=True, xticklabels=model.classes_, yticklabels=model.classes_)
```
