```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(copie_data, test_size=0.33, random_state=42)
print(f"Train set: ", train_set.shape)
print(f"Test set: ", test_set.shape)
```

```python
from sklearn.svm import SVC

x = train_set.drop('Classe', axis=1)
y = train_set['Classe']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

train_label_counts = y_train.value_counts()
train_total_count = train_label_counts.sum()
print(train_label_counts)
print(train_total_count)
train_toxic_percentage = (train_label_counts[1] / train_total_count) * 100
train_edible_percentage = (train_label_counts[0] / train_total_count) * 100

print(f"Pourcentage de champignons toxiques dans y_train: {train_toxic_percentage:.2f}%")
print(f"Pourcentage de champignons comestibles dans y_train: {train_edible_percentage:.2f}%")

test_label_counts = y_test.value_counts()
test_total_count = test_label_counts.sum()
test_toxic_percentage = (test_label_counts[1] / test_total_count) * 100
test_edible_percentage = (test_label_counts[0] / test_total_count) * 100

print(f"Pourcentage de champignons toxiques dans y_test: {test_toxic_percentage:.2f}%")
print(f"Pourcentage de champignons comestibles dans y_train: {test_edible_percentage:.2f}%")
```

```python
model = SVC(C=100)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, xticklabels=model.classes_, yticklabels=model.classes_)
```
