```python
data_first_review = data["first_review"]
data_first_review
# recupere que l'annÃ©e
data_first_review = pd.to_datetime(data_first_review)
data_first_review = data_first_review.dt.year
data_first_review

# 
df = pd.DataFrame(data["id"])
df["first_review"] = data_first_review
df["price"] = data["price"]
df = df.set_index("id")
df
# remplace les valeurs nulles par 2007
df = df.fillna(2007)
# enlÃ¨ve 2008 
df = df[df["first_review"] > 2008]
# voir la variation du prix avec seasborn
sns.lineplot(data=df, x="first_review", y="price")
```
