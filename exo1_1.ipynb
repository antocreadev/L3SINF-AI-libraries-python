```python
model = LinearRegression() # création du modèle
x = data_cars.drop(columns=["Consommation"]) # variable explicative
y = data_cars["Consommation"] # variable à expliquer
 
model.fit(x,y) # apprentissage
ypred = model.predict(x) # prédiction
model.score(x,y) # score de prédiction
```

```python
plt.scatter(x=y, y=ypred) 
# sur x on a les valeurs réelles et sur y les valeurs prédites
```

```python
# erreur quadratique : la différence entre les valeurs réelles et les valeurs prédites
sum((y - ypred )*(y - ypred)) # somme des erreurs quadratiques

# with mean_squared_error
mean_squared_error(y, ypred) # moyenne erreur quadratique moyenne
mean_squared_error(y, ypred) * data_cars.shape[0] # somme des erreurs quadratiques
```
